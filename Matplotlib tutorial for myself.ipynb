# 1. Importing Matplotlib

```python
import matplotlib.pyplot as plt
```

# 2. Basic Line Plot

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
plt.plot(x, y, label='Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Basic Line Plot')
plt.legend()
plt.show()
```

# 3. Scatter Plot

```python
plt.scatter(x, y, label='Scatter Plot', color='red', marker='o')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
plt.legend()
plt.show()
```

# 4. Bar Chart

```python
categories = ['Category 1', 'Category 2', 'Category 3']
values = [3, 7, 5]
plt.bar(categories, values, color='blue')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart')
plt.show()
```

# 5. Multi-Class Subplots

```python
plt.subplot(2, 2, 1)
plt.plot(x, y)
plt.title('Subplot 1')

plt.subplot(2, 2, 2)
plt.scatter(x, y, color='red', marker='o')
plt.title('Subplot 2')

plt.subplot(2, 2, 3)
plt.bar(categories, values, color='green')
plt.title('Subplot 3')

plt.subplot(2, 2, 4)
plt.pie(values, labels=categories, autopct='%1.1f%%', colors=['orange', 'purple', 'yellow'])
plt.title('Subplot 4')

plt.suptitle('Multi-Class Subplots')
plt.show()
```

# 6. Displaying Images

```python
img = plt.imread("image.png")
plt.imshow(img)
plt.axis('off')
plt.title('Image')
plt.show()
```

# 7. Multiple Images Subplot

```python
img1 = plt.imread("image.png")
img2 = plt.imread("image2.webp")

plt.subplot(1, 2, 1)
plt.imshow(img1)
plt.axis('off')
plt.title('Image 1')

plt.subplot(1, 2, 2)
plt.imshow(img2)
plt.axis('off')
plt.title('Image 2')

plt.suptitle('Multiple Images Subplot')
x=plt.show()
```
